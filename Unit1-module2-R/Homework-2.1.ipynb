### 3 Prime numbers

#### Find first 500 prime numbers. What is the 500th prime number?

```r
## Create a function that calculates first n prime numbers
## argument(s): integer n
## return value: a vector that contains the result

getPrimes <- function(n) {
    ## subfunction that decides whether a num is prime or not
    ## basically just iterate through prime list we already have
    ## possibly O(n^2) time complexity?
    
    decidePrime  <-  function(target, limit) {
        for (i in 1:limit) {
            ## we only need to examine prime that smaller than 
            ## square root of the target
            # text<-sprintf('targt is %s, i is %s, limit is %s \n', target, i, limit)
            # cat(text)
            if (res[i] > sqrt(target)) {
                return (TRUE);
            } else if (target %% res[i] == 0) {
                # fail fast
                return(FALSE)
            } 
        }
        
        return(TRUE)
    }

    ## init result vector
    res <- vector("integer", n)
    
    ## check invalid argument
    if (n < 1) {
        return(res)
    }
    
    ## check for special argument
    
    res[1] <- 2
    if (n == 1) {
        return(res)
    }
    
    ## find primes
    count <- 1
    cur <- 3
    while (count < n) {
        if (decidePrime(cur, count)) {
            res[count + 1] = cur;
            count = count + 1;
        }
        cur = cur + 1;
    }
    
    return(res)
   
    
}


```

```r
## calling the function
num <- 500
t1 <- Sys.time()
output <- getPrimes(num)
t2 <- Sys.time()
print(t2 - t1)
cat("the first 500 primes are: \n")
print(output[num])
```

### 4. Fibonacci number

#### Define a function to return the Xth fibonacci number

```r
## Recursively
## Time complexity: ~O(2^n)
## space complexity O(n) because of stack space used

getNthFibWithRecursion <- function(n) {
    if (n <= 2) {
        return(1);
    }
    
    # I guess we can just use a list to store all the temp result to save time
    return (getNthFibWithRecursion(n - 1) + getNthFibWithRecursion(n - 2))
}

getNthFibWithRecursion(30) # super slow, duh
```

```r
## Iteratively
## Time complexity O(n), which is way better than O(2^n)
## Space complexity O(1), just some constants

getNthFibWithIteration <- function(n) {
    if (n <= 2) {
        return(1);
    }
    
    first <- 1
    second <- 1
    
    sum <- 0
    for (i in 3:n) {
        sum <- first + second
        first <- second
        second <- sum
    }
    
    return(sum)
}

getNthFibWithIteration(30) # bit faster, huh?
```

#### 5.4 Define a function to evaluate padj output:
  - padj > 0.05: nonsig
  - padj <= 0.05: sig

```r
evalPadj <- function(input) {
    if (is.na(input)) {
        return (NA)
    }
    
    if (input > 0.05) {
        return("nonsig")
    }
    
    return ("sig")
}
```

#### 5.5 Apply the function to both nav.mem and nav.d14 and create one column to add the evaluation result.

```r
df.nav.d14$evalResult <- lapply(df.nav.d14$padj, evalPadj)
df.nav.mem$evalResult <- lapply(df.nav.mem$padj, evalPadj)
```
